# Task 1
#### Consider that 46% of the U.S. labor force is female and 25% of the females work part time. In addition, 17.4% of all American labourers are known to be part time workers. If a part time worker is randomly selected, what is the probability that the worker is a female?

P(W) = 46% = 0.46
P(T) = 17.4% = 0.174
P(T|W) = 25% = 0.25

P(W|T) = (P(W) * P(T|W)) / P(T)
       = (0.46 * 0.25) / 0.25
       = 0.115 / 0.25
       = 0.46
       = 46%

# Task 2
#### NIX Partners researched to determine the which changes in the office design would increase productivity. During the examination, 70% of respondents believed that reducing noise would increase productivity, and 67% said that more storage space would increase productivity. In addition, 56% of the respondents believed that reducing noise and increasing storage space will increase productivity. A worker is randomly selected and asked about the changes in the office design. The worker believes that reducing noise will increase productivity. What is the probability that the worker believes that more storage space will increase productivity?

For the non mutually exclusive events A and B:
Joint probability = P(A and B) = P(A ∩ B) = P(A) x P(B) P(A U B)

P(N) = 70% = 0.7
P(S) = 67% = 0.67
P(N and S) = 56% = 0.56
P(S | N) = P(S and N) / P(N)
           = 0.56 / 0.7
           = 0.8 = 80%
