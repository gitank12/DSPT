# d. Compute the mean and standard deviation of X. Interpret the mean in the context of the problem.

```python
import math
mean=0
for x,p in zip(A,Pa):
    mean=mean+x*p

var=0
for x,p in zip(A,Pa):
    var=var+((x-mean)**2)*p

std=math.sqrt(var)
print("Mean: ",mean)
print("Standard Deviation: ",std)
```

# Analysis
Std. Deviation is small, so the values are closely distributed with mean.

# a. What is the probability that no customers arrive in a 5 min interval time?

```python
import numpy as np
from scipy.stats import poisson

n1=[0]
rate= (3+6+4+6+2+3+1+5+1+0+3+3+1+2+4+0+2+6+5+4+2+5+3+4+5+3+5+3+5+4+7+3+4+8+3+1)/36
print(rate)
poisson1 = poisson.pmf(n1,rate)
print(poisson1)
```

```python
import matplotlib.pyplot as plt
plt.plot(n1,poisson1, 'o-')
plt.title('Poisson')
plt.xlabel('Number of customers arrived')
plt.ylabel('Prob of number of customers arrived')
plt.show()
```

# b. What is the probability that 6 or more customers arrive in a 5 min interval time?

```python
import numpy as np
from scipy.stats import poisson

n2=np.arange(6,20)
rate= (3+6+4+6+2+3+1+5+1+0+3+3+1+2+4+0+2+6+5+4+2+5+3+4+5+3+5+3+5+4+7+3+4+8+3+1)/36
poisson2=poisson.cdf(n2,rate) 
sum_p=0
for i in poisson2:
    sum_p=sum_p+i

print("Probability that 6 or more customers arrive in a 5 min interval time: ",sum_p)
```

```python
import matplotlib.pyplot as plt
plt.plot(n2,poisson2, 'o-')
plt.title('Poisson')
plt.xlabel('Number of customers arrived')
plt.ylabel('Prob of number of customers arrived')
plt.show()
```
